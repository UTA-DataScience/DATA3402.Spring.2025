```python
class Matrix:
    def __init__(self, data):
        if not all(len(row) == len(data[0]) for row in data):
            raise ValueError("All rows must have the same length.")
        self.data = data
        self.rows = len(data)
        self.columns = len(data[0])
    def __str__(self): 
        # String representation for printing the matrix
        return '\n'.join(['\t'.join(map(str, row)) for row in self.data])
    def shape(self):
        return (self.rows, self.columns)

    def transpose(self):
        # Returns a new Matrix instance which is the transpose of the original matrix
        transposed_data = [[self.data[j][i] for j in range(self.rows)] for i in range(self.columns)]
        return Matrix(transposed_data)

    def row(self, n):
        # Returns the nth row as a new Matrix instance.
        if n < 0 or n >= self.rows:
            raise IndexError("Row index out of range.")
        return Matrix([self.data[n]])  # Return as a new 1 row matrix

    def column(self, n):
        # Returns the nth column as a new Matrix instance.
        if n < 0 or n >= self.columns:
            raise IndexError("Column index out of range.")
        return Matrix([[self.data[i][n]] for i in range(self.rows)])  # Return as a new 1 column matrix

    def to_list(self):
        # Returns a list of lists (copy of the data)
        return [row[:] for row in self.data]

    def block(self, n_0, n_1, m_0, m_1):
        # Returns a new Matrix instance which is a copy of the specified block of the matrix
        if not (0 <= n_0 < n_1 <= self.rows) or not (0 <= m_0 < m_1 <= self.columns):
            raise IndexError("Invalid block indices.")
        
        submatrix = [row[m_0:m_1] for row in self.data[n_0:n_1]]
        return Matrix(submatrix)

    def __getitem__(self, key):
        # Overloading the [] operator for accessing matrix elements
        if isinstance(key, tuple):
            row_idx, col_idx = key
            if isinstance(row_idx, slice) and isinstance(col_idx, slice):
                # Slice both rows and columns
                sliced_data = [row[col_idx] for row in self.data[row_idx]]
                return Matrix(sliced_data)
            elif isinstance(row_idx, int) and isinstance(col_idx, slice):
                # Single row, multiple columns (returns as 1 row matrix)
                return Matrix([self.data[row_idx][col_idx]])
            elif isinstance(row_idx, slice) and isinstance(col_idx, int):
                # Multiple rows, single column (returns as 1 column matrix)
                return Matrix([[row[col_idx]] for row in self.data[row_idx]])
            elif isinstance(row_idx, int) and isinstance(col_idx, int):
                # Single element access
                return self.data[row_idx][col_idx]
            else:
                raise TypeError("Invalid index type for matrix slicing.")
        elif isinstance(key, int):
            # Access full row
            return Matrix([self.data[key]])  # Return as a 1 row matrix
        else:
            raise TypeError("Invalid index type.")
    
class Matrix_A(Matrix): # Where n is the number of rows and m is the number of columns
    def __init__(self, n, m):
        self.rows = n
        self.columns = m
        self.data = [[0 for _ in range(m)] for _ in range(n)] # List comprehension to create matrix with zeros
class Matrix_B(Matrix):
    def __init__(self, data):
        # Check if all rows have the same length
        if not all(len(row) == len(data[0]) for row in data):
            raise ValueError("All rows must have the same length.")
        self.data = data
        self.rows = len(data)
        self.columns = len(data[0])
```

```python
# Define 2x2 matrices
A = Matrix([[1, 2], [3, 4]])
B = Matrix([[2, 0], [1, 3]])
C = Matrix([[0, 1], [2, 1]])

# Define the 2x2 Identity Matrix
I = eye(2)

# Associativity of Matrix Multiplication: (AB)C == A(BC)
AB = A * B
BC = B * C
assoc_lhs = AB * C  # (AB)C
assoc_rhs = A * BC  # A(BC)

print("\nAssociativity of Matrix Multiplication: (AB)C == A(BC)")
print("LHS (AB)C:\n", assoc_lhs)
print("RHS A(BC):\n", assoc_rhs)
print("Equality Check:", assoc_lhs == assoc_rhs)

# Distributive Property: A(B + C) == AB + AC
B_plus_C = B + C
AB_plus_AC = A * B + A * C
distrib_lhs = A * B_plus_C  # A(B + C)
distrib_rhs = AB_plus_AC  # AB + AC

print("\nDistributive Property: A(B + C) == AB + AC")
print("LHS A(B + C):\n", distrib_lhs)
print("RHS AB + AC:\n", distrib_rhs)
print("Equality Check:", distrib_lhs == distrib_rhs)

# Non-Commutativity of Matrix Multiplication: AB â‰  BA
BA = B * A
print("\nNon-Commutativity: AB â‰  BA")
print("AB:\n", AB)
print("BA:\n", BA)
print("Equality Check:", AB == BA)

# Identity Matrix Property: AI = A
A_identity = A * I

print("\nIdentity Matrix Property: AI = A")
print("AI:\n", A_identity)
print("A:\n", A)
print("Equality Check:", A_identity == A)
```
